## Data Understanding
The yellow taxi trip records include fields capturing pick-up and drop-off dates/times, pick-up and drop-off locations, trip distances, itemized fares, rate types, payment types, and driver-reported passenger counts.

The data is stored in Parquet format (*.parquet*). The dataset is from 2009 to 2024. However, for this assignment, we will only be using the data from 2023.

The data for each month is present in a different parquet file. You will get twelve files for each of the months in 2023.

The data was collected and provided to the NYC Taxi and Limousine Commission (TLC) by technology providers like vendors and taxi hailing apps. <br>

You can find the link to the TLC trip records page here: https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page

#### Sampling the Data
> One way is to take a small percentage of entries for pickup in every hour of a date. So, for all the days in a month, we can iterate through the hours and select 5% values randomly from those. Use `tpep_pickup_datetime` for this. Separate date and hour from the datetime values and then for each date, select some fraction of trips for each of the 24 hours.

To sample data, you can use the `sample()` method. Follow this syntax:

```Python
# sampled_data is an empty DF to keep appending sampled data of each hour
# hour_data is the DF of entries for an hour 'X' on a date 'Y'

sample = hour_data.sample(frac = 0.05, random_state = 42)
# sample 0.05 of the hour_data
# random_state is just a seed for sampling, you can define it yourself

sampled_data = pd.concat([sampled_data, sample]) # adding data for this hour to the DF
```

This *sampled_data* will contain 5% values selected at random from each hour.

Note that the code given above is only the part that will be used for sampling and not the complete code required for sampling and combining the data files.

Keep in mind that you sample by date AND hour, not just hour. (Why?)

````python
# Take a small percentage of entries from each hour of every date.
# Iterating through the monthly data:
#   read a month file -> day -> hour: append sampled data -> move to next hour -> move to next day after 24 hours -> move to next month file
# Create a single dataframe for the year combining all the monthly data

# Select the folder having data files
import os

# Select the folder having data files
#os.chdir('/content/Assignments/EDA/data_NYC_Taxi/trip_records')
data_path = '/content/Datasets and Dictionary/trip_records/'

# Create a list of all the twelve files to read
file_list = [f for f in os.listdir(data_path) if f.endswith('.parquet')]

# initialise an empty dataframe
df = pd.DataFrame()


# iterate through the list of files and sample one by one:
for file_name in file_list:
    try:
        # file path for the current file
        file_path = os.path.join(data_path, file_name)

        # Reading the current file
        month_data = pd.read_parquet(file_path)

        # Extract date and hour from pickup datetime
        month_data['pickup_date'] = month_data['tpep_pickup_datetime'].dt.date
        month_data['pickup_hour'] = month_data['tpep_pickup_datetime'].dt.hour


        # We will store the sampled data for the current date in this df by appending the sampled data from each hour to this
        # After completing iteration through each date, we will append this data to the final dataframe.
        sampled_data = pd.DataFrame()

        # Loop through dates and then loop through every hour of each date
        for date in month_data['pickup_date'].unique():
            date_data = month_data[month_data['pickup_date'] == date]
            # Iterate through each hour of the selected date
            for hour in range(24):
                hour_data = date_data[date_data['pickup_hour'] == hour]
                # Sample 5% of the hourly data randomly
                if len(hour_data) > 0:
                    sample = hour_data.sample(frac=0.05, random_state=42)
                    sampled_data = pd.concat([sampled_data, sample])
                # add data of this hour to the dataframe
         # Drop the helper columns
        sampled_data = sampled_data.drop(columns=['pickup_date', 'pickup_hour'])

        # Append this month's sampled data to the final dataframe

        df = pd.concat([df, sampled_data])
        print(f"Processed {file_name}: {len(sampled_data)} rows sampled")
        # Concatenate the sampled data of all the dates to a single dataframe


    except Exception as e:
        print(f"Error reading file {file_name}: {e}")

    print(f"\nTotal rows in combined dataset: {len(df)}")
````

**3.2.1** <font color = red>[3 marks]</font> <br>
Identify slow routes by calculating the average time taken by cabs to get from one zone to another at different hours of the day.

Speed on a route *X* for hour *Y* = (*distance of the route X / average trip duration for hour Y*)

````python
import geopandas as gpd
# Find routes which have the slowest speeds at different times of the day
# Create a copy to avoid SettingWithCopy warnings
df_speed = df.copy()

# Ensure 'zones' GeoDataFrame is loaded
# The path to the shapefile should be from the unzipped data
shapefile_path = '/content/Datasets and Dictionary/taxi_zones/taxi_zones.shp'
zones = gpd.read_file(shapefile_path)

# 1. Calculate trip duration in hours
df_speed['duration_hours'] = (df_speed['tpep_dropoff_datetime'] - df_speed['tpep_pickup_datetime']).dt.total_seconds() / 3600

# 2. Calculate speed (mph)
# Filter for reasonable durations (e.g., > 1 min) to avoid division issues or parking trips
df_speed = df_speed[df_speed['duration_hours'] > (1/60)]
df_speed['speed_mph'] = df_speed['trip_distance'] / df_speed['duration_hours']

# 3. Categorize time of day
def get_time_period(hour):
    if 6 <= hour < 10: return 'Morning Rush (6-10)'
    elif 10 <= hour < 16: return 'Midday (10-16)'
    elif 16 <= hour < 20: return 'Evening Rush (16-20)'
    else: return 'Late Night/Early Morning'

df_speed['time_period'] = df_speed['tpep_pickup_datetime'].dt.hour.apply(get_time_period)

# 4. Group by Route (Pick-up to Drop-off) and Time Period
# We filter for routes with at least 10 trips to ensure the 'slow speed' is a pattern, not a one-off
route_speeds = df_speed.groupby(['PULocationID', 'DOLocationID', 'time_period']).agg(
    avg_speed=('speed_mph', 'mean'),
    trip_count=('speed_mph', 'count')
).reset_index()

# Filter for meaningful data
route_speeds = route_speeds[route_speeds['trip_count'] >= 5]

# 5. Merge zone names for readability (Optional but helpful)
# Assuming 'zones' GeoDataFrame has 'LocationID' and 'zone'
zone_map = zones[['LocationID', 'zone']].set_index('LocationID')['zone'].to_dict()

route_speeds['pickup_zone'] = route_speeds['PULocationID'].map(zone_map)
route_speeds['dropoff_zone'] = route_speeds['DOLocationID'].map(zone_map)

# 6. Find the 10 slowest routes across different times
slowest_routes = route_speeds.sort_values(by='avg_speed').head(15)

print("Top 15 Slowest Routes (Average Speed < 5 mph often indicates heavy traffic):")
slowest_routes[['pickup_zone', 'dropoff_zone', 'time_period', 'avg_speed', 'trip_count']]
````

##### Pricing Strategy

**3.2.9** <font color = red>[2 marks]</font> <br>
For the different passenger counts, find the average fare per mile per passenger.

For instance, suppose the average fare per mile for trips with 3 passengers is 3 USD/mile, then the fare per mile per passenger will be 1 USD/mile.

````python
# Analyse the fare per mile per passenger for different passenger counts

# Create a copy and filter out zero distance/passengers
df_metrics = df[(df['trip_distance'] > 0) & (df['passenger_count'] > 0)].copy()

# 1. Calculate Fare per Mile per Passenger
# This represents the "cost efficiency" for the passengers
df_metrics['fare_per_mile_per_person'] = df_metrics['fare_amount'] / (df_metrics['trip_distance'] * df_metrics['passenger_count'])

# 2. Group by Passenger Count and calculate stats
# We calculate the median to be robust against outliers (e.g., short trips with high base fares)
efficiency_stats = df_metrics.groupby('passenger_count')['fare_per_mile_per_person'].agg(['mean', 'median', 'count']).reset_index()

# 3. Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=efficiency_stats, x='passenger_count', y='median', palette='coolwarm')

plt.title('Median Fare per Mile per Passenger', fontsize=16, fontweight='bold')
plt.xlabel('Number of Passengers', fontsize=12)
plt.ylabel('Cost ($) per Mile per Person', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Add value labels
for i, row in efficiency_stats.iterrows():
    plt.text(i, row['median'] + 0.1, f"${row['median']:.2f}", ha='center', fontweight='bold')

plt.show()

# Summary
print("Summary Statistics for Cost Efficiency:")
print(efficiency_stats)

print("\nInsight: The cost per person per mile decreases significantly as more passengers share the cab,")
print("making large group rides (5-6 passengers) much more economical per individual.")

````

Find the average fare per mile by hours of the day and by days of the week

````python
# Compare the average fare per mile for different days and for different times of the day
# Calculate Fare per Mile and compare across Day and Time

# 1. Setup Time Periods and Day Types
df_fare_mile = df[df['trip_distance'] > 0.5].copy() # Filter for > 0.5 miles to avoid base-fare skew
df_fare_mile['fare_per_mile'] = df_fare_mile['fare_amount'] / df_fare_mile['trip_distance']

# Reuse time period categories
def get_time_period(hour):
    if 7 <= hour < 10: return 'Morning Rush (7-10 AM)'
    elif 10 <= hour < 16: return 'Midday (10 AM-4 PM)'
    elif 16 <= hour < 19: return 'Evening Rush (4-7 PM)'
    elif 19 <= hour < 23: return 'Late Evening (7-11 PM)'
    else: return 'Night (11 PM-7 AM)'

df_fare_mile['time_period'] = df_fare_mile['tpep_pickup_datetime'].dt.hour.apply(get_time_period)
df_fare_mile['day_type'] = df_fare_mile['tpep_pickup_datetime'].dt.dayofweek.apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# 2. Aggregate
fare_pivot = df_fare_mile.pivot_table(
    index='time_period',
    columns='day_type',
    values='fare_per_mile',
    aggfunc='median'
)

# Reorder index for logical time flow
time_order = ['Night (11 PM-7 AM)', 'Morning Rush (7-10 AM)', 'Midday (10 AM-4 PM)', 'Evening Rush (4-7 PM)', 'Late Evening (7-11 PM)']
fare_pivot = fare_pivot.reindex(time_order)

# 3. Plot Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(fare_pivot, annot=True, fmt=".2f", cmap="YlOrRd", cbar_kws={'label': 'Median Fare per Mile ($)'})

plt.title('Median Fare per Mile by Time and Day Type', fontsize=16, fontweight='bold')
plt.ylabel('Time Period', fontsize=12)
plt.xlabel('Day Category', fontsize=12)

plt.show()

# Insight
print("Insight: Fare per mile tends to be higher during rush hours due to congestion surcharges")
print("and slower travel speeds, while night fares (where traffic is light) often show lower per-mile costs")
print("unless the trip involves long highway distances at higher speeds.")

````
